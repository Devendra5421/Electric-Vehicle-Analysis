```python
# Re-import necessary libraries and reload the data
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV files again
ev_state = pd.read_csv('electric_vehicle_sales_by_state.csv', encoding='UTF-8-SIG')

# Convert date to datetime and extract year
ev_state['date'] = pd.to_datetime(ev_state['date'], format='%d-%b-%y')
ev_state['year'] = ev_state['date'].dt.year

# Filter data for 4-Wheelers in 2023 and 2024
ev_4w_2023_2024 = ev_state[(ev_state['vehicle_category'] == '4-Wheelers') & (ev_state['year'].isin([2023, 2024]))]

# Group by year and sum the sales
yearly_4w_sales = ev_4w_2023_2024.groupby('year')['electric_vehicles_sold'].sum()

print("4-Wheeler EV Sales in 2023 and 2024:")
print(yearly_4w_sales)

# Visualize the sales trends
plt.figure(figsize=(8, 5))
yearly_4w_sales.plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('4-Wheeler EV Sales in 2023 and 2024')
plt.xlabel('Year')
plt.ylabel('Total 4-Wheeler EV Sales')
plt.xticks(rotation=0)
plt.tight_layout()
#plt.savefig('4w_ev_sales_2023_2024.png')
plt.show()

print("\
4-Wheeler EV Sales in 2023 and 2024 chart saved as '4w_ev_sales_2023_2024.png'")
```
